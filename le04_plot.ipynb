# Sinais discretos
Desafio: Plote os seguintes sinais discretos

- $ y_1(t) = 2 \sin(2\pi f t) $
- $ y_2(t) = \sin(2\pi f t) + \sin(4\pi f t) $
- $ y_3(t) = \frac{4}{\pi} \sin(2\pi f t) + \frac{4}{3\pi} \sin(6\pi f t) $
- $ y_4(t) = y_3(t) + \frac{4}{5\pi} \sin(10\pi f t) + \frac{4}{7\pi} \sin(14\pi f t) + \frac{4}{9\pi} \sin(18\pi f t) $

# ![sines.png](https://github.com/2023-2-INF0413-PDSI/materiais/blob/main/le04_sinais-digitais/sines.png?raw=1)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def signal_y1(t, f):
  return 2 * np.sin(2 * np.pi * f * t)

def signal_y2(t, f):
  return np.sin(2 * np.pi * f * t) + np.sin(4 * np.pi * f * t)

def signal_y3(t, f):
  return (4 / np.pi) * np.sin(2 * np.pi * f * t) + (4 / (3 * np.pi)) * np.sin(6 * np.pi * f * t)

def signal_y4(t, f):
  return signal_y3(t, f) + (4/(5*np.pi))*np.sin(10 * np.pi * f * t) + (4/(7*np.pi))*np.sin(14 * np.pi * f * t) + (4/(9*np.pi))*np.sin(18 * np.pi * f * t)
```

```python
t = np.linspace(0, 0.1, 1000)
frequency = 20
```

```python
# Y1

y1 = signal_y1(t, frequency)

plt.figure(figsize=(8, 4))

plt.plot(t, y1)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('y_1(t)')
plt.show()
```

```python
# Y2

y2 = signal_y2(t, frequency)

plt.figure(figsize=(8, 4))
plt.plot(t, y2)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('y_2(t)')
plt.show()
```

```python
# Y3

y3 = signal_y3(t, frequency)

plt.figure(figsize=(8, 4))
plt.plot(t, y3)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('y_3(t)')
plt.show()
```

```python
# Y4

y4 = signal_y4(t, frequency)

plt.plot(t, y4)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('y_4(t)')
plt.show()
```
